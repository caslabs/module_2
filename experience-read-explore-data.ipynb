## Data Munging Process

We convert records.txt input file to Panda's DataFrame data structure to utilize data analysis operations

```python
path = "records.txt"
records = [json.loads(line) for line in open(path)]
frame = DataFrame(records)
frame
```

Now, to convert the top time zones into Windows and non-Windows user - we must make sure the dataset does not contain any NaN (missing values) within the agents data. We clean our data by creating another frame that excludes NaN using notnull() operation.

```python
cframe = frame[frame.a.notnull()]
cframe
```

Now, we categorize the data to Windows and non-Windows user. We utilize numpy where() operation to compute for each a value is Windows or not with str.contains() operation.

```python
operating_system = np.where(cframe['a'].str.contains('Windows'), 'Windows', 'Not Windows')
operating_system
```

We then group the data by its time zone column - by_tz_os frame is now a frame that consists of {timezone, not windows, and windows}

```python
by_tz_os = cframe.groupby(['tz', operating_system])
```

We then compute the group counts using the size() operation. Then we unstack() to reshape it to make a table then we clean it further by filling na values to 0.

```python
agg_counts = by_tz_os.size().unstack().fillna(0)
agg_counts[:10]
```

To select the top overall time zones, we create a indirect index array that rom the row count and sort by ascending order.

```python
indexer = agg_counts.sum(1).argsort()
indexer[:10]
```

Then we slice off the last 10 rows (pythonic slicing notation) and using the take() operation.

```python
count_subset = agg_counts.take(indexer)[-10:]
count_subset
```

We can visualize count_subset dataframe as a bar plot with stacked=True option to replicate Figure 2-3.

```python
count_subset.plot(kind='barh', stacked=True)
```

Lastly, we must normalized the row to sum to 1 - such that we can see the relative percentage of Windows users in the smaller groups. Now our plot is the replication of Figure 2-3.

```python
normed_subset = count_subset.div(count_subset.sum(1), axis=0)
normed_subset.plot(kind='barh', stacked=True)
```
